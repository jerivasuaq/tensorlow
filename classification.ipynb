```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sigmoid(z:float):
    return 1/(1+np.exp(-z))
```

```python
z = np.linspace(-10,10,100)
a = sigmoid(z)
```

```python
plt.plot(z, a)
```

```python
from sklearn.datasets import make_blobs
```

```python
features, labels = make_blobs(n_samples=10, n_features=2, centers=2, random_state=75)
plt.scatter(features[:,0], features[:,1], c=labels, cmap='coolwarm')
```

```python

print('f0 min', min(features[:,0]))
print('f0 max', max(features[:,0]))
print('f1 min', min(features[:,1]))
print('f1 max', max(features[:,1]))


print(features[:,0])
print(labels)
```

```python
x0 = (features[:,0]-min(features[:,0]))/(max(features[:,0])- min(features[:,0]))
x1 = (features[:,1]-min(features[:,1]))/(max(features[:,1])- min(features[:,1]))
print(x0)
print(x1)
plt.scatter(x0, x1, c=labels, cmap='coolwarm')
```

```python
x = np.linspace(0,11,10)
y = -x + 5
plt.scatter(features[:,0], features[:,1], c=labels, cmap='coolwarm')
plt.plot(x,y)
```

## [1,1] * F - 5 = 0 

```python
F = [[8],[10]]
np.array([1,1]).dot(np.array(F)) - 5
```

```python
F = [[2],[-10]]
np.array([1,1]).dot(np.array(F)) - 5
```

```python
from tensorlow.nodes import Graph, Placeholder, Variable
from tensorlow.session import Session
from tensorlow.operations import add, matmul, Sigmoid
```

```python
g = Graph()
g.set_as_default()
x = Placeholder() 
w = Variable([1,1])
b = Variable(-5)
z = add(matmul(w,x),b)
a = Sigmoid(z)
```

```python
sess = Session()
sess.run(operation=a, feed_dict={x: [8,10]})
```

```python
sess.run(operation=a, feed_dict={x: [2,-10]})
```
